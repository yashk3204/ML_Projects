Data Preprocessing

```python
from google.colab import files
import pandas as pd

obj = files.upload()
df = pd.read_csv(list(obj.keys())[0])
df.head()
```

```python
df = pd.DataFrame(df)
df
```

```python
# Checking for missing values
print(df.isnull().sum())
```

```python
df.columns
```

```python
# Separating the DF into x and y (features and target)
y = df['Outcome']
x = df
x = df.drop(['Outcome'], axis=1)
```

```python
# Splitting data into training (80%) and testing (20%) sets
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# Feature Scaling using Standard Scaler
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

Naïve Bayes Model

```python
from sklearn.naive_bayes import GaussianNB
naive_bayes_model = GaussianNB()
naive_bayes_model.fit(x_train,y_train)
```

```python
y_pred_nb=naive_bayes_model.predict(x_test)
y_pred_nb
```

```python
# Calculating accuracy, precision, recall and F1 score
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score

accuracy = accuracy_score(y_test, y_pred_nb)
precision = precision_score(y_test, y_pred_nb, average='weighted')
recall = recall_score(y_test, y_pred_nb, average='weighted')
f1 = f1_score(y_test, y_pred_nb, average='weighted')

print(f'Accuracy: {accuracy}\nPrecision: {precision}\nRecall: {recall}\nF1 Score: {f1}')
```

Perceptron

```python
from sklearn.linear_model import Perceptron
perceptron_model = Perceptron (max_iter=1000, tol=1e-3, random_state=42)
perceptron_model.fit(x_train, y_train)
```

```python
y_pred_perceptron = perceptron_model.predict(x_test)
y_pred_perceptron
```

```python
# Calculating accuracy, precision, recall and F1 score

accuracy = accuracy_score(y_test, y_pred_perceptron)
precision = precision_score(y_test, y_pred_perceptron, average='weighted')
recall = recall_score(y_test, y_pred_perceptron, average='weighted')
f1 = f1_score(y_test, y_pred_perceptron, average='weighted')

print(f'Accuracy: {accuracy}\nPrecision: {precision}\nRecall: {recall}\nF1 Score: {f1}')
```

The Naïve Bayes model outperforms the Perceptron model in all metrics - accuracy, precision, recall and F1 score. It predicts positive instances better while predicting less false positives. It also has a higher recall, so it predicts less false negatives, which is crucial in datasets such as this where we want to predict diseases.

Implementing K-Fold Cross Validation with 8 folds to evaluate model performances

```python
from sklearn.model_selection import cross_val_score, KFold
kf = KFold(n_splits=8, shuffle=True, random_state=1)
```

```python
# For the Naive Bayes model
scores = cross_val_score(naive_bayes_model, x, y, cv=kf)
print(f"Cross-validation scores for each fold: {scores}")
print(f"Mean accuracy: {scores.mean()}")
print(f"Standard deviation of accuracy: {scores.std()}")
```

```python
# For the perceptron model
scores = cross_val_score(perceptron_model, x, y, cv=kf)
print(f"Cross-validation scores for each fold: {scores}")
print(f"Mean accuracy: {scores.mean()}")
print(f"Standard deviation of accuracy: {scores.std()}")
```

The Naive Bayes model has a higher accuracy on average across the 8 folds and also has a lower standard deviation. Hence it is more accurate and more consistent and is the better, more reliable choice.

Custom Implementation of Perceptron:

```python
import numpy as np

class Perceptron:

    def __init__(self, learning_rate, n_iters):
        self.learning_rate = learning_rate
        self.n_iters = n_iters
        self.weights = None
        self.bias = None

    def fit(self, x, y):
        n_samples, n_features = x.shape
        np.random.seed(56)
        self.weights = np.random.randn(n_features)
        self.bias = np.random.rand()
        y_ = np.where(y <= 0, -1, 1) # Converting y labels from {0, 1} to {-1, 1} as Perceptron uses a sign function
        for _ in range(self.n_iters):
            for idx, xi in enumerate(x):
                linear_output = np.dot(xi, self.weights) + self.bias
                y_predicted = np.sign(linear_output)
                if y_predicted != y_[idx]:
                    self.weights += self.learning_rate * y_[idx] * xi
                    self.bias += self.learning_rate * y_[idx]

    def predict(self, x):
        linear_output = np.dot(x, self.weights) + self.bias
        return np.sign(linear_output)
```

```python
perceptron = Perceptron(learning_rate=0.05, n_iters=1000)
perceptron.fit(x_train, y_train)
y_pred_custom_perceptron = perceptron.predict(x_test)
```

```python
# Redefining how to calculate accuracy as we changed the labels of y from {0, 1} to {-1, 1}
def print_accuracy(y_test, y_pred):
  y_test = np.array(y_test)
  y_pred = np.array(y_pred)
  tp = np.sum((y_pred == 1) & (y_test == 1))
  tn = np.sum((y_pred == -1) & (y_test == -1))
  fp = np.sum((y_pred == 1) & (y_test == -1))
  fn = np.sum((y_pred == -1) & (y_test == 1))
  accuracy = (tp + tn) / (tp + tn + fp + fn)
  print(accuracy)

print_accuracy(y_test, y_pred_custom_perceptron)
```
