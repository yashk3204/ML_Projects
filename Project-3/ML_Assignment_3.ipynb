```python
#Separating the DF into x and y (features and target) and dropping the serial no. column
y = df['Chance of Admit ']
x = df
x = df.drop(['Chance of Admit ', 'Serial No.'], axis=1)
```

```python
# Binning i.e. discretizing the target variable into three categories
import numpy as np
bins = np.percentile(y, [33, 67])  # Calculating 33rd and 67th percentiles
labels = ['Low', 'Medium', 'High']
y_binned = pd.cut(y, bins=[-float('inf'), bins[0], bins[1], float('inf')], labels=labels, include_lowest=True)
print(y_binned)
```

```python
# Applying minmax scaler to every feature of x
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x = pd.DataFrame(scaler.fit_transform(x), columns=x.columns)
```

We use Min-Max scaler in regression problems such as this to preserve the relationship between features. It improves model performance as it helps gradient descent algorithms converge faster. Also in LDA, it helps as it treats all features equally.

Splitting and scaling is done

Pre-processing done.

```python
# Using the categorized target variable to perform Linear Discriminant Analysis (LDA) on the input features.
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
lda = LinearDiscriminantAnalysis()
x_lda = lda.fit_transform(x, y_binned)
print(x_lda)
```

The number of components retained after LDA is 2, i.e. one less than the number of classes. The components are ranked based on the amount of between-class variance they explain. LDA maximizes class separability and reduces the risk of overfitting, by retaining the most important components.

```python
# Converting the LDA-transformed data into TensorFlow tensors
import tensorflow as tf
x_lda_tensor = tf.convert_to_tensor(x_lda, dtype=tf.float32)
print(x_lda_tensor)
```

```python
import tensorflow as tf

# Initialize parameters
W = tf.Variable(tf.random.normal([x_lda_tensor.shape[1], 1]))  # Weights (shape [2, 1] for 2 features)
b = tf.Variable(tf.random.normal([1]))  # Bias (shape [1])
learning_rate = 0.01
epochs = 1000

# Linear regression function
def linear_regression(X):
    return tf.matmul(X, W) + b

# Mean Squared Error (MSE) Loss function
def mse_loss(y_pred, y_true):
    return tf.reduce_mean(tf.square(y_pred - y_true))

# Ensuring y has correct shape [500, 1]
y_tensor = tf.convert_to_tensor(y, dtype=tf.float32)
y_tensor = tf.reshape(y_tensor, (-1, 1))  # Reshaping y to [500, 1]

# Training loop without GradientTape
for epoch in range(epochs):
    y_pred = linear_regression(x_lda_tensor)
    loss = mse_loss(y_pred, y_tensor)
    error = y_pred - y_tensor
    N = tf.cast(tf.shape(x_lda_tensor)[0], tf.float32)
    dW = (2 / N) * tf.matmul(tf.transpose(x_lda_tensor), error)
    db = (2 / N) * tf.reduce_sum(error)
    db = tf.reshape(db, b.shape)
    W.assign_sub(learning_rate * dW)
    b.assign_sub(learning_rate * db)
    # Finding loss after every 100 epochs
    if (epoch + 1) % 100 == 0:
        print(f'Epoch [{epoch+1}/{epochs}], Loss: {loss.numpy()}')
```

```python
import matplotlib.pyplot as plt

# Assuming y_pred contains predicted values
y_pred = linear_regression(x_lda_tensor).numpy()

# Converting y_tensor to a NumPy array
y_true = y_tensor.numpy()

# Creating a scatter plot of actual vs. predicted values
plt.scatter(y_true, y_pred)
plt.xlabel("Actual Labels")
plt.ylabel("Predicted Labels")
plt.title("Actual vs. Predicted Labels")
plt.plot([min(y_true), max(y_true)], [min(y_true), max(y_true)], color='red', linestyle='--') # Adding a diagonal line for comparison
plt.show()
```
